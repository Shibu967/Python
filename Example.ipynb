```python
#time series plot()
import matplotlib.pyplot as plt
import datetime
import numpy as np
import pandas as pd
df=pd.DataFrame({'date':np.array([datetime.datetime(2020,1,(i+1)) for i in range(10)]),"umang":[1,2,4,3,5,6,7,25,1,9]})
plt.plot(df.date,df.umang,label='hours',linewidth=0.6,linestyle="--",color='m')
plt.legend()
plt.title("Time Series plot")
plt.xlabel("date")
plt.ylabel("No of hours")
plt.show()
```

# labels,Annotation and legends

```python
import matplotlib.pyplot as plt
values1=[5,8,9,4,1,6,7,2,3,8]
values2=[8,3,2,7,6,1,4,9,8,5]
plt.plot(range(1,11),values1)
plt.plot(range(1,11),values2)
plt.xlabel("Roll no")
plt.ylabel("CPI")
plt.annotate(xy=[5,1],s="Lowest CPI")
plt.legend(['cx','cy'],loc=5)
plt.show()
```
